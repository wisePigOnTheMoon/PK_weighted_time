```python
import csv
import matplotlib.pyplot as plt 

with open('weighted-to-timeline.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
x=[]
y=[]

print(data[1][3])
print(len(data[1][3]))

for i in range(len(data)-1):
    row=data[i+1]
    try:
        if len(row[3])==11:
            if int(row[3][4]+row[3][5])>25:
                y.append(float(row[7]))
                x.append(int("19"+row[3][4]+row[3][5]))
                
            else:
                y.append(float(row[7]))
                x.append(int("20"+row[3][4]+row[3][5]))
                
                
        elif len(row[3])==12:
            if int(row[3][5]+row[3][6])>25:
                y.append(float(row[7]))
                x.append(int("19"+row[3][5]+row[3][6]))
                
            else:
                y.append(float(row[7]))
                x.append(int("20"+row[3][5]+row[3][6]))
                

        elif len(row[3])==12:
            if int(row[3][5]+row[3][6])>25:
                y.append(float(row[7]))
                x.append(int("19"+row[3][5]+row[3][6]))
                
            else:
                y.append(float(row[7]))
                x.append(int("20"+row[3][5]+row[3][6]))
                
        else:
            print("failed")
    except:
        print("error")
    
    
    if len(x) !=len(y):
        print(x)
        print(y)
        break


plt.plot(x, y,'.') 
plt.xlabel('years') 
# naming the y axis 
plt.ylabel('admet lab weighted score') 
  
# giving a title to my graph 
plt.title('year vs weighted score') 
  
# function to show the plot 
plt.show()
```

For each year determine the mean drug weighted score and graph that vrs years

```python
import numpy as np


data_dict = {}
for year, value in zip(x, y):
    if year not in data_dict:
        data_dict[year] = [value]
    else:
        data_dict[year].append(value)


# graphing mean values
mean_values = {year: np.mean(values) for year, values in data_dict.items()}
plt.scatter(list(mean_values.keys()), list(mean_values.values()), label='Data Points', marker='o')

#  line of best fit
coefficients = np.polyfit(list(mean_values.keys()), list(mean_values.values()), 1)
poly = np.poly1d(coefficients)
x_fit = np.linspace(min(mean_values.keys()), max(mean_values.keys()), 100)
y_fit = poly(x_fit)
plt.plot(x_fit, y_fit, label='Line of Best Fit', color='red')

plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.title('Year vs Mean drug score')
plt.legend()
plt.show()
```

Graph a boxplot displaying more information such as median, q1, q3, and outliers in each year

```python
data_dict = {}
for year, value in zip(x, y):
    if year not in data_dict:
        data_dict[year] = [value]
    else:
        data_dict[year].append(value)

fig, ax = plt.subplots(figsize=(20, 30))
box_plot_data = [values for year, values in data_dict.items()]
ax.boxplot(box_plot_data, labels=list(data_dict.keys()), vert=True)      #***** horizontal ***
#ax.boxplot(box_plot_data, labels=list(data_dict.keys()))                 #**** VERTICAL

ax.set_xlabel('Values')
ax.set_title('Box Plot of Values for Each Year')
plt.savefig('Figures/box_plot_vertical-2.png')
plt.show()
```
